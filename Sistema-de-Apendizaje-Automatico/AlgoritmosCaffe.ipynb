Algoritmo 1: KNN clasifica un dato nuevo basándose en los ejemplos más cercanos a él. Busca los “k vecinos” más próximos y asigna la clase más frecuente.

```python
# Importo todo lo necesario.
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# 1. Cargar dataset
data = load_iris()
X, y = data.data, data.target

# 2. Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Crear modelo KNN
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# 4. Predicción y precisión
y_pred = knn.predict(X_test)
acc_knn = accuracy_score(y_test, y_pred)

print("Precisión KNN:", acc_knn)
```

Alogitmo 2: SVM encuentra la mejor línea o hiperplano que separa las clases con el mayor margen posible.

```python
# Algoritmo 2: SVM Independiente
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# 1. Cargar dataset
data = load_iris()
X, y = data.data, data.target

# 2. Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 3. Crear modelo SVM
svm = SVC(kernel='linear')
svm.fit(X_train, y_train)

# 4. Predicción y precisión
y_pred = svm.predict(X_test)
acc_svm = accuracy_score(y_test, y_pred)

print("Precisión SVM:", acc_svm)
```

Este algoritmo divide los datos en ramas mediante reglas simples (si/entonces) hasta llegar a una predicción final.

```python
# Algoritmo 3: Árbol de Decisión Independiente
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# 1. Cargar dataset
data = load_iris()
X, y = data.data, data.target

# 2. Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 3. Crear modelo Árbol de Decisión
tree = DecisionTreeClassifier(max_depth=3)
tree.fit(X_train, y_train)

# 4. Predicción y precisión
y_pred = tree.predict(X_test)
acc_tree = accuracy_score(y_test, y_pred)

print("Precisión Árbol de Decisión:", acc_tree)
```
