Insertar registros
Vuelve a conectarte a la base de datos escuela.db y crea un cursor.
Inserta al menos tres registros en la tabla estudiantes. Utiliza sentencias INSERT con marcadores de posición ? para evitar inyecciones SQL.
Guarda los cambios y cierra la conexión.

```python
# Importamos las librerias necesarias
import sqlite3

# Conectamos a la base de datos (o la creamos si no existe)
conn = sqlite3.connect('escuela.db')

# Creamos un cursor para ejecutar comandos SQL
cursor = conn.cursor()

# Insertamos registros en la tabla de estudiantes.
cursor.execute("INSERT INTO estudiantes (nombre, edad, nota) VALUES ('Ana', 20, 8.5)")
cursor.execute("INSERT INTO estudiantes (nombre, edad, nota) VALUES ('Luis', 22, 9.0)")
cursor.execute("INSERT INTO estudiantes (nombre, edad, nota) VALUES ('Marta', 21, 7.5)")

# Guardamos los cambios y cerramos la conexión
conn.commit()
conn.close()
```
