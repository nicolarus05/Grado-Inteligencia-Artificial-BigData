Crear una base de datos y una tabla
Conecta a una base de datos SQLite llamada escuela.db usando sqlite3.connect().
Crea un cursor.
Define una tabla llamada estudiantes con los siguientes campos:
id como entero y clave primaria.
nombre como texto.
edad como entero.
nota como real (float).
Guarda (comita) los cambios y cierra la conexión.

```python
# Importamos las librerias necesarias
import sqlite3 as db

# Conectamos a la base de datos (o la creamos si no existe)
conn = db.connect('escuela.db')

# Creamos un cursor para ejecutar comandos SQL
cursor = conn.cursor()

# Creamos la tabla 'estudiantes' si no existe
cursor.execute('''
    CREATE TABLE IF NOT EXISTS estudiantes (
        id INTEGER PRIMARY KEY,
        nombre TEXT NOT NULL,
        edad INTEGER,
        nota FLOAT
    )
''')

# Confirmamos los cambios y cerramos la conexión
conn.commit()
conn.close()
```
