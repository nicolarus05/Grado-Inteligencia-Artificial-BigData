Manejo de errores
Escribe un bloque de código try/except que intente crear de nuevo la tabla estudiantes en la base de datos escuela.db. Como la tabla ya existe, deberías atrapar el error que genere SQLite.
Muestra un mensaje personalizado informando de que la tabla ya existe cuando se capture la excepción.
Recuerda cerrar la conexión en un bloque finally si la conexión se abrió correctamente.

```python
# Importamos las librerias necesarias
import sqlite3

# Conectamos a la base de datos (o la creamos si no existe)
conn = sqlite3.connect('escuela.db')

# Creamos un cursor para ejecutar comandos SQL
cursor = conn.cursor()

# Volvemos a crear la tabla estudiantes
try:
    cursor.execute('''
        CREATE TABLE estudiantes (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            nombre TEXT NOT NULL,
            edad INTEGER,
            grado TEXT
        )
    ''')
    print("Tabla 'estudiantes' creada exitosamente.")
except sqlite3.OperationalError as e:
    print("La tabla 'estudiantes' ya existe. No se creó de nuevo.")
    print("Error:", e)

# Cerramos la conexión a la base de datos
conn.close()
```
