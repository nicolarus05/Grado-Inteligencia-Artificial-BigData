Usar pandas con SQLite
Tras haber insertado varios registros en la base de datos, conecta a escuela.db.
Utiliza la función pandas.read_sql_query() para leer todos los registros de la tabla estudiantes en un DataFrame.
Muestra el DataFrame resultante y calcula la edad media de los estudiantes usando el método mean() sobre la columna edad.
Cierra la conexión.

```python
# Importamos las librerias necesarias
import pandas as pd
import sqlite3

# Conectamos a la base de datos SQLite
conn = sqlite3.connect('escuela.db')

# Leemos la tabla de estudiantes
df_estudiantes = pd.read_sql_query("SELECT * FROM estudiantes", conn)

# Mostramos los datos de los estudiantes
print(df_estudiantes)

# Calculamos la edad media de los estudiantes
edad_media = df_estudiantes['edad'].mean()
print(f'La edad media de los estudiantes es: {edad_media}')

# Cerramos la conexión a la base de datos
conn.close()
```
