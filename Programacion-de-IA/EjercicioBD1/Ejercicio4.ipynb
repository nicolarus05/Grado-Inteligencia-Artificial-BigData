Actualizar y borrar registros
Conéctate a la base de datos y crea un cursor.
Actualiza la nota de uno de los estudiantes (elige un id concreto) utilizando una sentencia UPDATE.
Borra otro estudiante de la tabla utilizando una sentencia DELETE por su id.
Consulta nuevamente la tabla completa para comprobar los cambios.
Guarda los cambios y cierra la conexión.

```python
# Importamos las librerias necesarias
import sqlite3

# Conectamos a la base de datos (o la creamos si no existe)
conn = sqlite3.connect('escuela.db')

# Creamos un cursor para ejecutar comandos SQL
cursor = conn.cursor()

# Actualizamos la nota del segundo estudiante.
cursor.execute("UPDATE estudiantes SET nota = ? WHERE id = ?", (9.5, 2))
conn.commit()
print("Nota actualizada correctamente.")

# Borramos el tercer estudiante.
cursor.execute("DELETE FROM estudiantes WHERE id = ?", (3,))
conn.commit()
print("Estudiante borrado correctamente.")

# Consultamos todos los estudiantes para verificar los cambios
cursor.execute("SELECT * FROM estudiantes")
estudiantes = cursor.fetchall()
for estudiante in estudiantes:
    print(estudiante)

# Cerramos la conexión a la base de datos
conn.close()
```
