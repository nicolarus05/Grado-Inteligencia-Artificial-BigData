Ejercicio 1: Contar palabras en un archivo de texto
Escribe un programa que lea un archivo de texto y cuente cuántas veces aparece cada palabra (ignora mayúsculas y signos de puntuación). El resultado debe almacenarse en un diccionario donde las claves sean las palabras y los valores sus respectivas frecuencias. Al final, muestra las palabras y sus conteos ordenados alfabéticamente.

Sugerencias:
Utiliza el método lower() para convertir todas las palabras a minúsculas.
Puedes eliminar signos de puntuación con el módulo string o usando expresiones regulares.
Recorre cada línea del archivo y divide el texto en palabras usando split().

```python
# Importamos las librerias para poder leer el CSV
import pandas as pd
import string

# Nombre del archivo CSV
archivo = 'Ejercicio1.csv'

# Creamos el diccionario para almacenar las palabras y sus conteos
conteo_palabras = {}

# Abrimos el archivo CSV
with open(archivo, 'r', encoding='utf-8') as f:
    lector = pd.read_csv(f, header=None)
    
    # Iteramos sobre cada fila del CSV
    for index, fila in lector.iterrows():
        for celda in fila:
            # Convertimos el texto a minusculas.
            texto = str(celda).lower()
            
            # Eliminamos signos de puntuacion
            texto = texto.translate(str.maketrans('', '', string.punctuation))
            
            # Dividimos el texto en palabras
            palabras = texto.split()
            
            # Contamos las palabras
            for palabra in palabras:
                conteo_palabras[palabra] = conteo_palabras.get(palabra, 0) + 1

# Mostramos los resultados ordenados alfabeticamente
for palabra, conteo in sorted(conteo_palabras.items()):
    print(f'{palabra}: {conteo}')
```
