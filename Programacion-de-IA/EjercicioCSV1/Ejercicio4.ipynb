Media de ventas mensuales con pandas
Tienes un CSV ventas.csv con dos columnas: Fecha y Ventas, donde Fecha contiene fechas en formato AAAA-MM-DD y Ventas son números enteros. Utiliza pandas para calcular la media de ventas por mes y mostrar el resultado en un DataFrame.

Sugerencias:

Carga el archivo con pandas.read_csv() y convierte la columna Fecha al tipo datetime con pd.to_datetime().
Usa groupby() para agrupar por el año y el mes. Puedes usar df.groupby(df['Fecha'].dt.to_period('M')).
Emplea mean() para obtener la media de ventas por mes.
Opcionalmente, reinicia el índice para mostrar la columna de período como una columna normal.

```python
# Importamos las librerias necesarias
import pandas as pd

# Cargamos el archivo usando pandas.read_csv
archivo = pd.read_csv('ventas.csv')

# Convertimos la columna fecha a tipo datetime
archivo['fecha'] = pd.to_datetime(archivo['Fecha'], format='%Y-%m-%d')

# Usamos groupby para agrupar por el año y el mes.
ventas_mensuales = archivo.groupby(archivo['fecha'].dt.to_period('M'))

# Usamos mean para obtener la media de ventas por mes
media_ventas_mensuales = ventas_mensuales['Ventas'].mean()

print(media_ventas_mensuales)
```
