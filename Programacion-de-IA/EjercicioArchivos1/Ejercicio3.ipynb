Filtrar datos de un CSV
Dispones de un archivo CSV llamado personas.csv con los campos nombre, edad y ciudad. Escribe un programa que lea este archivo y cree uno nuevo llamado mayores_de_30.csv que contenga solo las filas de las personas cuya edad sea mayor que 30. Utiliza el módulo csv de la biblioteca estándar.

Sugerencias:

Usa csv.DictReader para leer cada fila como un diccionario.
Al escribir el nuevo archivo, utiliza csv.DictWriter y asegúrate de escribir la fila de encabezado (writeheader()).
Convierte el campo de edad a entero para comparar correctamente.

```python
# Importamos el csv
import csv

# Leemos el archivo personas.csv usando csv.DictReader
with open('personas.csv', 'r', encoding='utf-8') as archivo_entrada:
    lector = csv.DictReader(archivo_entrada)
    
    # Filtramos las personas mayores de 30 años
    mayores_de_30 = [fila for fila in lector if int(fila['edad']) >= 30]

# Escribimos el nuevo archivo usando csv.DictWriter
with open('mayores_de_30.csv', 'w', newline='', encoding='utf-8') as archivo_salida:
    # Definimos los nombres de las columnas
    campos = ['nombre', 'edad', 'ciudad']
    escritor = csv.DictWriter(archivo_salida, fieldnames=campos)
    
    # Escribimos el encabezado
    escritor.writeheader()
    
    # Escribimos las filas filtradas
    escritor.writerows(mayores_de_30)
```
