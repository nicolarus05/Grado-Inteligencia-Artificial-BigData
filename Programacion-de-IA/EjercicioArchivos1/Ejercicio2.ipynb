Copiar un archivo de texto
Crea una función copiar_archivo(origen, destino) que reciba dos nombres de archivo y copie el contenido del archivo de origen al archivo de destino. Asegúrate de manejar correctamente la apertura y cierre de los archivos utilizando with.

Sugerencias:

Abre el archivo de origen en modo lectura ('r') y el de destino en modo escritura ('w').
Lee el contenido del archivo de origen línea a línea o de una sola vez y escríbelo en el archivo de destino.
Comprueba al final que el archivo de destino contiene exactamente el mismo texto que el archivo de origen.

```python
# Creamos la funcion copiar_archivo
def copiar_archivo(origen, destino):
    # Abrimos el archivo de origen en modo lectura
    with open(origen, 'r') as archivo_origen:
        # Leemos el contenido del archivo de origen
        contenido = archivo_origen.read()
        
    # Abrimos el archivo de destino en modo escritura
    with open(destino, 'w') as archivo_destino:
        # Escribimos el contenido en el archivo de destino
        archivo_destino.write(contenido)
        
# Ejemplo de uso
copiar_archivo('Ejercicio1.csv', 'Copia_Ejercicio1.csv')
```
